```python
df = pd.read_csv('diamonds.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.columns.values
```

```python
plt.figure(figsize= (100,45))
sns.relplot(x='carat',y='price',hue='clarity',alpha=.5,data=df)
plt.xlabel('Weight')
plt.ylabel('Price')
```

###### Concentration Gradient of Diamond Qualities


#  Mix by Weight and Cut 

```python
sns.displot(data=df,
            x="carat",
            hue="cut",
            kind="kde",
            height=6,
            multiple="fill",
            clip=(0, None),
            palette="ch:start=.1,rot=-.1")
plt.xlabel('Weight')
```

# Sales price by amont sold

```python
plt.figure(figsize=(10,5))
sns.histplot(x='price',hue='color',multiple="stack",edgecolor=".3",linewidth=.5,
    log_scale=True,palette='mako',data=df)
plt.xlabel('Price')
```

# Price by clarity 

```python
plt.figure(figsize=(10,6))
sns.barplot(x='clarity',y='price',palette='coolwarm',data=df)
plt.xlabel('Clarity')
plt.ylabel('Price')
```

The Distribution of Diamond colors by depth

```python
plt.figure(figsize=(20,10))
plt.figure(figsize=(10,5))
sns.histplot(x='depth',hue='color',multiple="stack",edgecolor=".3",linewidth=.5,
    log_scale=False,palette='cubehelix',data=df)
plt.xlim(55,70)
plt.xlabel('Depth')
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x='table',hue='color',multiple="stack",edgecolor=".3",linewidth=.5,
    log_scale=False,palette='cubehelix',data=df)
plt.xlim(50,70)
plt.xlabel('width of top of diamond relative to widest point')
```
